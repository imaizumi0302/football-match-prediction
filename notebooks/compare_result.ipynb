```python
df = pd.read_csv(data_file)

display(df)
```

```python
def best_metric(data,metric):
    if metric == "log_loss": 
        data = data.sort_values(metric,ascending=True)
    else:
        data = data.sort_values(metric,ascending=False)
    # 'Params_JSON' 列の文字列をすべてPythonの辞書に変換し、新しい列に格納
    data['Params_Dict'] = data['Params_JSON'].apply(json.loads)
    best_data = data.iloc[0,:]

    best_features = best_data["Features"]
    best_params = best_data["Params_Dict"]
    best_metric = best_data[metric]
    
    print("best_features")
    print(best_features)
    print()
    print("best_parameter")
    print(best_params)
    print()
    print("best_metric({})".format(metric))
    print(best_metric)
    print()
    display(data.head(1).T)
    
```

```python
best_metric(df,"log_loss") #確認したいmetricの種類を指定
```
